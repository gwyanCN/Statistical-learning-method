### 随机生成数据
#### 感知机属于误分类驱动,每次对每个分类点进行分类处理，通过误分类误差来学习调整模型

```python
#随机生成30个随机点每个点
data = np.random.randint(2,30,[2,15])
data2 =np.random.randint(30,60,[2,15])
data=np.hstack([data,data2])
labels = np.ones(30)*-1
labels[15:]=1
data = data/data.max()#归一化
plt.figure(figsize=(10,6))
plt.scatter(data[0][:15],data[1][:15],color="r")
plt.scatter(data[0][15:],data[1][15:],color="b")
plt.title("random Data")
plt.show()
data.shape
```

```python
## 初始化参数w,b
# 不同的初始化随机位置 也会使得迭代的次数与变化不同 
weight = np.zeros([2,1])
bias = np.zeros(1)
def classifer(weight, bias,x):
    return np.sign(np.dot(weight.T,x)+bias)
```

```python
# 绘制初始线
epsilon=10e-10
a1,a2=data[1].min(),data[1].max()
plt.scatter(data[0][:15],data[1][:15],color="r")
plt.scatter(data[0][15:],data[1][15:],color="b")
# ax1=plt.plot([a1,a2],[(-weight[0]/(weight[1]+epsilon))*a1-bias/(weight[1]+epsilon),(-weight[0]/(weight[1]+epsilon))*a2-bias/(weight[1]+epsilon)],'black')
while True:
    OUT = True
    for i in range(data.shape[1]):
        y = labels[i]
        while(y*classifer(weight,bias,data[:,i])<=0):
            OUT=False
            weight +=0.8*y*data[:,i].reshape(-1,1)
            bias +=0.8*y
        plt.scatter(data[0][:15],data[1][:15],color="r")
        plt.scatter(data[0][15:],data[1][15:],color="b")
        ax1=plt.plot([a1,a2],[(-weight[0]/(weight[1]+epsilon))*a1-bias/(weight[1]+epsilon),(-weight[0]/(weight[1]+epsilon))*a2-bias/(weight[1]+epsilon)],'black') 
        plt.show()
    if OUT:
        break
```
